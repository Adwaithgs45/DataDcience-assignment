# ***Diabetes Dataset EDA***

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv("/content/diabetes_model.csv")
df
````

Univariate Analysis

````python
plt.hist(df['BMI'],bins=10, color='Yellow', edgecolor='black')
plt.xlabel('BMI')
plt.ylabel('Range')
plt.title('BMI Range')
plt.show()
````

The amount of people is marked along the y-axis with the particular BMI value marked in x-axis.There is more amount of people with the BMI value around 20-45 having values from 75-200 and gradually keeps the value below 75 for all other values till the end.

````python
plt.hist(df['Insulin'],bins=10, color='Black', edgecolor='Red')
plt.xlabel('Insulin')
plt.ylabel('Range')
plt.title('Insulin Range')
plt.show()
````

The amount of people is marked along the y-axis with the particular insulin value marked in x-axis.
There is more amount of people with theinsulin value around 0-200 having values from 380-50 and gradually keeps the value below 50 for all other values till the end.


Bivariate Analysis


````python
sns.scatterplot(x='Pregnancies', y='BloodPressure',data=df, color="red", edgecolor="black")
plt.title('Scatter Plot of Pregnancies VS Blood Pressure')
plt.xlabel('Pregnancies')
plt.ylabel('BloodPressure')
plt.show()
````

The above graph shows points on which both the number of prgnancies and Blood Pressure of the these people.
This scatterplot shows no clear relationship between the number of pregnancies and blood pressure levels in this dataset. There is no clear upward or downward trend in blood pressure as the number of pregnancies increases.

 Multivariate Analysis

````python
sns.pairplot(df[['Age','Glucose','BMI']])
plt.show()
````

The plot involves three variables: Age, Glucose, and BMI.
The diagonal shows histograms for each variable. These histograms represent the distribution of each variable individually. For example, the top-left histogram represents the distribution of Age, the middle histogram represents Glucose, and the bottom-right histogram represents BMI.
The off-diagonal elements are scatter plots. Each scatter plot shows the relationship between two variables. For instance:
The top-right scatter plot shows Age on the x-axis and Glucose on the y-axis.
The bottom-left scatter plot shows Age on the x-axis and BMI on the y-axis.
The middle scatter plot (between Glucose and BMI) shows their relationship.
